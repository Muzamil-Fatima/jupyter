Implementation of activation functions in python

Sigmoid

```python
import math

def segmoid(x):
    return 1/ (1 + math.exp(-x))
```

```python
segmoid(100)
```

```python
segmoid(1)
```

```python
segmoid(-56)
```

```python
segmoid(0.5)
```

#tanh

```python
def tanh(x):
    return (math.exp(x) - math.exp(-x)) / (math.exp(x) + math.exp(-x)) 
```

```python
tanh(-56)
```

```python
tanh(50)
```

```python
tanh(1)
```

```python
def relu(x):
    return max(0, x)
```

```python
relu(8)
```

```python
relu(-8)
```

```python
relu(-100)
```

Leaky ReLU

```python
def leaky_relu(x):
    return max(0.1*x, x)
```

```python
leaky_relu(-100)
```

```python
leaky_relu(8)
```
